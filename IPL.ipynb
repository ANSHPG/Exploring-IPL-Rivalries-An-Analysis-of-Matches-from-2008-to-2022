```python
rr_point = summary['RAJASTHAN'].sum()
dc_point = summary['DELHI'].sum()
```

```python
plt.figure(facecolor='#121212')#121212
plt.title('                     VICTORY% [MATCH < 10 & JAIPUR], ANSHP-CHARTS',color='white')
# plt.xlabel('ANSHP CHARTS',color='white')
colors=['#00f3bb','#5367ff']
colors2=['#cd5855','#b2b2b2','']
label_colors=['#00f3bb','#5367ff']
pie = plt.pie(point,colors=colors,autopct='%1.1f%%',startangle=0)
plt.legend(labels=point_labels, loc='best', bbox_to_anchor=(1, 0.5))
for text in pie[1]:
    text.set_color(label_colors[0])
for text in pie[2]:
    text.set_color('black')
```
